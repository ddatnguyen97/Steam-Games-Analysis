```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def calculate_pct_na_values(df):
    return df.isna().sum() / len(df) * 100

for column in df.columns:
    print(f"{column}: {calculate_pct_na_values(df[column])}")
```

```python
df.columns
```

```python
columns_to_drop = ['About the game',
                    'Full audio languages', 
                    'Reviews', 
                    'Header image', 
                    'Website',
                    'Support email',
                    'Support url',
                    'Metacritic score', 
                    'Metacritic url',
                    'Average playtime two weeks',
                    'Median playtime two weeks',
                    'Median playtime forever',
                    'Screenshots',
                    'Movies',
                    'Score rank',
                    'Notes',
                    'Discount',
                    'User score',
                    'Recommendations',
                    ]

df = df.drop(columns=columns_to_drop)
df.columns
```

```python
df['Release date'] = pd.to_datetime(df['Release date'], format='mixed')
df['Release month'] = pd.to_datetime(df['Release date']).dt.to_period('M')
df['Release year'] = pd.to_datetime(df['Release date']).dt.year
df['Release year'] = df['Release year'].astype('int64')
```

```python
df.info()
```

```python
df.dropna(subset=['Name'], axis=0, inplace=True)
df.shape
```

```python
keywords = ['playtest', 'beta', 'alpha', 'sdk', 'server']
pattern = '|'.join(keywords)
df.drop(df[df['Name'].str.contains(pattern, case=False)].index, inplace=True)
df.shape
```

```python
def create_os_column(df):
    for index, row in df.iterrows():
        if row['Windows'] == 1:
            df.at[index, 'Supported OS'] = 'Windows'
            if row['Mac'] == 1:
                df.at[index, 'Supported OS'] += ', Mac'
                if row['Linux'] == 1:
                    df.at[index, 'Supported OS'] += ', Linux'
            elif row['Linux'] == 1:
                df.at[index, 'Supported OS'] += ', Linux'
        elif row['Mac'] == 1:
            df.at[index, 'Supported OS'] = 'Mac'
            if row['Linux'] == 1:
                df.at[index, 'Supported OS'] += ', Linux'
        elif row['Linux'] == 1:
            df.at[index, 'Supported OS'] = 'Linux'
        else:
            df.at[index, 'Supported OS'] = 'Other'
    return df

df = create_os_column(df)
```

```python
df.drop(columns=['Windows', 'Mac', 'Linux'], inplace=True)
df.head(10)
```

```python
def create_fop_column(df):
    if df['Price'] == 0:
        return 'Free'
    else:
        return 'Paid'
    
df['Free or paid'] = df.apply(create_fop_column, axis=1)
df.head(10)
```

```python
df['Estimated owners']
```

```python
def create_median_owned_column(range_str):
    if '-' in range_str:
        low, high = range_str.split(' - ')
        return (int(low) + int(high)) / 2
    return 0

df['Median estimated owners'] = df['Estimated owners'].apply(create_median_owned_column)
df.drop(columns=['Estimated owners'], inplace=True)
df.head(10)
```

```python
median_positive_reviews = df['Positive'].median()
print(f"Median positive reviews: {median_positive_reviews}")

print(df[df['Positive'] < median_positive_reviews].shape)
print(df[df['Positive'] >= median_positive_reviews].shape)
```

```python
median_negative_reviews = df['Negative'].median()
print(f"Median negative reviews: {median_negative_reviews}")

print(df[df['Negative'] < median_negative_reviews].shape)
print(df[df['Negative'] >= median_negative_reviews].shape)
```

```python
def calculate_sentiment_score(row):
    if row['Positive'] == 0 & row['Negative'] == 0:
        return -1
    elif row['Positive'] == 0 & row['Negative'] > 0:
        return 0
    else:
        return row['Positive'] / (row['Positive'] + row['Negative'])

df['Sentiment score'] = df.apply(calculate_sentiment_score, axis=1)
df.head(10)
```

```python
def create_sentiment_category(score):
    if 0.8 <= score <= 1: 
        return 'Very Positive'
    elif 0.6 <= score < 0.8:
        return 'Positive'
    elif 0.4 <= score < 0.6:
        return 'Neutral'
    elif 0.2 <= score < 0.4:
        return 'Negative'
    elif 0 <= score < 0.2: 
        return 'Very Negative'
    else:
        return 'No Reviews'
    
df['Sentiment category'] = df['Sentiment score'].apply(create_sentiment_category)
df.head(10)
```

```python
df.dropna(subset=['Categories', 'Tags', 'Genres', 'Publishers', 'Developers'], axis=0, inplace=True)
```

```python
cols_to_process = ['Genres', 'Categories', 'Tags']

for col in cols_to_process:
    df[col] = df[col].str.split(',')

df.head(10)
```
